### ANOVA

Generamos pruebas ANOVA para los cisticercos en cada tiempo de T0 a T3

```python
import scipy.stats as stats

# Extraer los valores de cisticercos para cada grupo en cada tiempo
T0_S3Pvac = S3Pvac["Cisticercos_T0"]
T0_Wild = Wild["Cisticercos_T0"]
T0_Salina = Salina["Cisticercos_T0"]

T1_S3Pvac = S3Pvac["Cisticercos_T1"]
T1_Wild = Wild["Cisticercos_T1"]
T1_Salina = Salina["Cisticercos_T1"]

T2_S3Pvac = S3Pvac["Cisticercos_T2"]
T2_Wild = Wild["Cisticercos_T2"]
T2_Salina = Salina["Cisticercos_T2"]

# Realizar ANOVA para cada tiempo
anova_T0 = stats.f_oneway(T0_S3Pvac, T0_Wild, T0_Salina)
anova_T1 = stats.f_oneway(T1_S3Pvac, T1_Wild, T1_Salina)
anova_T2 = stats.f_oneway(T2_S3Pvac, T2_Wild, T2_Salina)

# Mostrar resultados
print("Resultados ANOVA:")
print(f"T0: F={anova_T0.statistic:.4f}, p-value={anova_T0.pvalue:.4f}")
print(f"T1: F={anova_T1.statistic:.4f}, p-value={anova_T1.pvalue:.4f}")
print(f"T2: F={anova_T2.statistic:.4f}, p-value={anova_T2.pvalue:.4f}")

# Interpretación
alpha = 0.05
for i, anova in enumerate([anova_T0, anova_T1, anova_T2], start=0):
    tiempo = f"T{i}"
    if anova.pvalue < alpha:
        print(f"En {tiempo}, hay diferencias significativas entre los grupos (p < {alpha}).")
    else:
        print(f"En {tiempo}, NO hay diferencias significativas entre los grupos (p >= {alpha}).")
```

Generamos pruebas ANOVA para las lesiones en cada tiempo de T0 a T3


```python
# Extraer los valores de lesiones para cada grupo en cada tiempo
T0_S3Pvac = S3Pvac["Lesiones_T0"]
T0_Wild = Wild["Lesiones_T0"]
T0_Salina = Salina["Lesiones_T0"]

T1_S3Pvac = S3Pvac["Lesiones_T1"]
T1_Wild = Wild["Lesiones_T1"]
T1_Salina = Salina["Lesiones_T1"]

T2_S3Pvac = S3Pvac["Lesiones_T2"]
T2_Wild = Wild["Lesiones_T2"]
T2_Salina = Salina["Lesiones_T2"]

# Realizar ANOVA para cada tiempo
anova_T0 = stats.f_oneway(T0_S3Pvac, T0_Wild, T0_Salina)
anova_T1 = stats.f_oneway(T1_S3Pvac, T1_Wild, T1_Salina)
anova_T2 = stats.f_oneway(T2_S3Pvac, T2_Wild, T2_Salina)

# Mostrar resultados
print("Resultados ANOVA para Lesiones:")
print(f"T0: F={anova_T0.statistic:.4f}, p-value={anova_T0.pvalue:.4f}")
print(f"T1: F={anova_T1.statistic:.4f}, p-value={anova_T1.pvalue:.4f}")
print(f"T2: F={anova_T2.statistic:.4f}, p-value={anova_T2.pvalue:.4f}")

# Interpretación
alpha = 0.05
for i, anova in enumerate([anova_T0, anova_T1, anova_T2], start=0):
    tiempo = f"T{i}"
    if anova.pvalue < alpha:
        print(f"En {tiempo}, hay diferencias significativas entre los grupos (p < {alpha}).")
    else:
        print(f"En {tiempo}, NO hay diferencias significativas entre los grupos (p >= {alpha}).")
```

# **Análisis de Componentes Principales (PCA)**

El **Análisis de Componentes Principales (PCA)** es una técnica estadística que permite reducir la dimensionalidad de un conjunto de datos, transformando variables originales en un nuevo sistema de coordenadas llamado **componentes principales**, los cuales son ortogonales entre sí.

## **Procedimiento Matemático**

Dado un conjunto de datos con $n$ observaciones y $p$ variables, los pasos para calcular PCA son los siguientes:

### **1. Estandarización de los datos**
Para cada variable $X_j$, restamos su media para centrar los datos alrededor del origen:

$$
X'_{ij} = X_{ij} - \bar{X_j}
$$

### **2. Cálculo de la matriz de covarianza**
La matriz de covarianza $\Sigma$ se define como:

$$
\Sigma = \frac{1}{n-1} X'^T X'
$$

donde cada elemento $\sigma_{ij}$ de la matriz representa la covarianza entre las variables $X_i$ y $X_j$.

### **3. Cálculo de autovalores y autovectores**
Se resuelve la ecuación característica:

$$
\det(\Sigma - \lambda I) = 0
$$

para obtener los autovalores $\lambda_1, \lambda_2, \dots, \lambda_p$ y sus respectivos autovectores $v_1, v_2, \dots, v_p$.

### **4. Selección de componentes principales**
Se ordenan los autovalores de mayor a menor, y se seleccionan los $k$ primeros que expliquen al menos un porcentaje de varianza deseado. La proporción de varianza explicada por cada componente $i$ se calcula como:

$$
\frac{\lambda_i}{\sum \lambda_j}
$$

### **5. Proyección de los datos**
Finalmente, los datos originales son proyectados en el nuevo espacio de componentes utilizando la matriz de autovectores seleccionados $V_k$:

$$
Z = X' V_k
$$

Este procedimiento nos permite reducir la dimensionalidad de los datos mientras se conserva la mayor cantidad de información posible, asi que lo generamos dados nuestros 3 tipos de vacunas

## S3Pvac

```python
# Seleccionar solo las columnas numéricas para PCA en S3Pvac
datos_s3pvac = S3Pvac[['Cisticercos_T0', 'Cisticercos_T1', 'Cisticercos_T2', 'Lesiones_T0', 'Lesiones_T1', 'Lesiones_T2']]
datos_s3pvac_centrados = datos_s3pvac - datos_s3pvac.mean()
matriz_cov_s3pvac = np.cov(datos_s3pvac_centrados.T)
autovalores_s3pvac, autovectores_s3pvac = np.linalg.eig(matriz_cov_s3pvac)
indices = np.argsort(autovalores_s3pvac)[::-1]
autovalores_s3pvac = autovalores_s3pvac[indices]
autovectores_s3pvac = autovectores_s3pvac[:, indices]
componentes_s3pvac = np.dot(datos_s3pvac_centrados, autovectores_s3pvac[:, :2])
S3Pvac["Componente_1"] = componentes_s3pvac[:, 0]
S3Pvac["Componente_2"] = componentes_s3pvac[:, 1]


# Calcular la varianza explicada por cada componente
varianza_explicada_s3pvac = autovalores_s3pvac / np.sum(autovalores_s3pvac)
varianza_acumulada_s3pvac = np.cumsum(varianza_explicada_s3pvac)

# Mostrar la varianza explicada por cada componente
for i, (ve, va) in enumerate(zip(varianza_explicada_s3pvac, varianza_acumulada_s3pvac), 1):
    print(f"Componente {i}: Varianza explicada = {ve:.4f}, Varianza acumulada = {va:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.bar(range(1, len(varianza_explicada_s3pvac) + 1), varianza_explicada_s3pvac * 100, alpha=0.7, color="blue", label="Varianza Explicada")
plt.plot(range(1, len(varianza_explicada_s3pvac) + 1), varianza_acumulada_s3pvac * 100, marker='o', linestyle='--', color="red", label="Varianza Acumulada")
plt.xlabel("Número de Componentes")
plt.ylabel("Porcentaje de Varianza Explicada")
plt.title("Varianza Explicada por Componentes Principales (S3Pvac)")
plt.legend()
plt.grid()
plt.show()
```

Como podemos observar

```python
# Graficar los resultados
plt.figure(figsize=(8, 6))
plt.scatter(S3Pvac["Componente_1"], S3Pvac["Componente_2"], color='red', label='Observaciones')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.title("Proyección de los datos en el espacio de los componentes principales")
plt.legend()
plt.grid()
plt.show()
```

El Análisis de Componentes Principales (PCA) se utilizó en los tres grupos experimentales (S3Pvac-Papaya12mg, Papaya Wild Type y Solución Salina) con dos enfoques diferentes:

PCA manual (usando álgebra matricial con autovalores y autovectores).

En ambos métodos, se llega a la misma conclusión: solo un componente principal es suficiente para explicar casi toda la variabilidad en los datos.
